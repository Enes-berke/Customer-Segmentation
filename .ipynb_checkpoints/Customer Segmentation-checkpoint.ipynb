## 3. Veri Görselleştirme

````python
# Tarih formatını dönüştürme ve eksik değerleri temizleme
data["InvoiceDate"] = pd.to_datetime(data["InvoiceDate"],errors = 'coerce')
data = data.dropna(subset=["InvoiceDate"])
````

````python
# 1. Fatura başına toplam ürün sayısı

data["InvoiceYearMonth"] = data["InvoiceDate"].apply(lambda x : x.strftime('%y-%m')) 
# Tarihleri Yıl-Ay formatında bir string olarak oluşturur.

invoice_count = data.groupby("InvoiceYearMonth")["InvoiceNo"].count()
#Her Yıl-Ay kombinasyonu için fatura sayısını hesaplar.

plt.figure(figsize=(12,6))
plt.plot(invoice_count)
plt.title('Aylık Fatura Sayısı')
plt.xlabel('Yıl-Ay')
plt.ylabel('Fatura Sayısı')
plt.grid(True)
plt.show()
````

````python
# 2. Fiyat dağılımı

plt.figure(figsize=(12,6))
sns.histplot(data["UnitPrice"], bins = 20, kde = True)

plt.xlim(0,7000)
plt.ylim(0,2500)

plt.title("Fiyat Dağılımı")
plt.xlabel("Fiyat")
plt.ylabel("Frekans")
plt.grid(True)
plt.show()
````

````python
# 3. Ürün kategorilerine göre satış
top_products = data["Description"].value_counts().head(25) # En çok satan 25 ürün

plt.figure(figsize=(12,6))
sns.barplot(x=top_products.index, y=top_products.values)
plt.title('En Çok Satılan 10 Ürün')
plt.xlabel('Ürünler')
plt.ylabel('Satış Miktarı')
plt.xticks(rotation=90)
plt.show()
````

````python
# 4. Ülkelere göre satış

country_sales = data.groupby("Country")["InvoiceNo"].count().sort_values(ascending=False)

plt.figure(figsize=(12,6))
sns.barplot(x=country_sales.index[:10], y=country_sales.values[:10])
plt.title('Ülkelere Göre Satış')
plt.xlabel('Ülkeler')
plt.ylabel('Fatura Sayısı')
plt.xticks(rotation=90)
plt.show()
````

````python
# 5. Fatura başına toplam harcama

data['TotalPrice'] = data['Quantity'] * data['UnitPrice'] # Her ürünün toplam fiyatını hesaplar
invoice_total = data.groupby('InvoiceNo')['TotalPrice'].sum() # Her fatura için toplam harcamayı hesaplar.

plt.figure(figsize=(12,6))
sns.histplot(invoice_total, bins=50, kde=True)
plt.title('Fatura Başına Toplam Harcama Dağılımı')
plt.xlabel('Toplam Harcama')
plt.ylabel('Frekans')
plt.grid(True) # Izgara çizgilerini görünür hale getirir
plt.show()
````

````python
# 6. En çok alışveriş yapan ülkeler

plt.figure(figsize=(12,6))

data['TotalPrice'] = data['Quantity'] * data['UnitPrice']
country_sales = data.groupby('Country')['TotalPrice'].sum().sort_values(ascending=False).head(10)

sns.barplot(x=country_sales.index, y=country_sales.values)
plt.title('Ülkelere Göre Toplam Satışlar')
plt.xlabel('Ülkeler')
plt.ylabel('Toplam Satışlar')
plt.xticks(rotation=90)
plt.show()
````

````python
# 7. Zaman bazlı satış trendi (haftalık)

weekly_sales = data.set_index('InvoiceDate').resample('W')['TotalPrice'].sum()

# Veriyi InvoiceDate üzerine zaman serisi olarak yeniden örnekler ve haftalık toplam satışları hesaplar.

plt.figure(figsize=(12,6))
plt.plot(weekly_sales)
plt.title('Haftalık Satış Trendleri')
plt.xlabel('Tarih')
plt.ylabel('Toplam Satışlar')
plt.grid(True)
plt.show()
````

* Aylık Fatura Sayısı: Zaman içindeki satış trendlerini anlamak için fatura sayısını zaman bazında çizdik.
* Fiyat Dağılımı: Ürün fiyatlarının dağılımını görmek için histogram oluşturduk.
* En Çok Satılan Ürünler: En çok satılan ürünleri bar grafikle gösterdik.
* Ülkelere Göre Satış: Satışların hangi ülkelerde yoğunlaştığını görmek için bir bar grafik kullandık.
* Fatura Başına Toplam Harcama: Müşterilerin fatura başına ne kadar harcadıklarını histogram ile analiz ettik.
* Ülkelere Göre Toplam Satışlar: Satışların hangi ülkelerde parasal olarak yoğunlaştığını inceledik.
* Haftalık Satış Trendleri: Zaman serisi analizi yaparak haftalık satışları çizdik.

## 4. Veriyi Modele Hazırlama

````python
# RFM analizi yaparak müşterileri segmente edebilmek için veriyi dönüştüreceğiz. 
# Bu aşamada her müşterinin Recency, Frequency ve Monetary (RFM) skorlarını hesaplayacağız.
````

````python
# Recency (En son alışveriş tarihi)
import datetime as dt
snapshot_date = data['InvoiceDate'].max() + dt.timedelta(days=1)

rfm = data.groupby('CustomerID').agg({
    'InvoiceDate': lambda x: (snapshot_date - x.max()).days,  # Recency
    'InvoiceNo': 'nunique',  # Frequency
    'TotalPrice': 'sum'  # Monetary
})

rfm.columns = ['Recency', 'Frequency', 'Monetary']

# Monetary değerinde 0 ve negatif olan müşterileri çıkarıyoruz
rfm = rfm[rfm['Monetary'] > 0]

print(rfm.head(25))
````

Bu kod, RFM (Recency, Frequency, Monetary) analizi yapmak için yazılmıştır. RFM analizi, müşteri verilerini anlamak ve segmentlere ayırmak amacıyla kullanılan bir teknik olup, müşterilerin davranışlarını değerlendirmek için üç temel metriği kullanır: **Recency (Son Alışveriş)**, **Frequency (Sıklık)** ve **Monetary (Harcamalar)**. Şimdi kodun her bir bölümünü detaylıca açıklayalım:

### 1. İthalat ve Tarih Hesaplama

```python
import datetime as dt
snapshot_date = data['InvoiceDate'].max() + dt.timedelta(days=1)
```

- **`import datetime as dt`**: `datetime` modülünü `dt` olarak içe aktarır. Bu modül tarih ve zaman ile ilgili işlemleri yapmanıza olanak sağlar.
- **`snapshot_date`**: Bu değişken, veri setindeki en son fatura tarihini alır ve bir gün ekleyerek hesaplar. Bu tarih, müşterilerin en son alışveriş tarihini referans alarak gelecekteki analizler için bir "kesme noktası" oluşturur. Örneğin, en son fatura tarihi 2024-01-15 ise, `snapshot_date` 2024-01-16 olur.

### 2. RFM Analizi

```python
rfm = data.groupby('CustomerID').agg({
    'InvoiceDate': lambda x: (snapshot_date - x.max()).days,  # Recency
    'InvoiceNo': 'nunique',  # Frequency
    'TotalPrice': 'sum'  # Monetary
})
```

- **`data.groupby('CustomerID')`**: Müşteri ID'sine göre veri grubuna ayırır. Yani, her müşteri için ayrı bir grup oluşturur.
- **`agg({...})`**: Her müşteri grubu için toplu işlemler yapar. Burada üç farklı agregat fonksiyon kullanılır:

  - **`'InvoiceDate': lambda x: (snapshot_date - x.max()).days`**: Her müşteri için son alışveriş tarihini `snapshot_date` ile karşılaştırarak, `Recency` (Son Alışveriş) hesaplanır. `x.max()` en son fatura tarihini alır. `(snapshot_date - x.max()).days` ifadesi, gün cinsinden son alışveriş tarihine olan uzaklığı hesaplar.
  - **`'InvoiceNo': 'nunique'`**: Her müşteri için fatura sayısının (yani alışveriş sayısının) benzersizliğini hesaplar. Bu `Frequency` (Sıklık) metriği sağlar.
  - **`'TotalPrice': 'sum'`**: Her müşteri için toplam harcamaları hesaplar. Bu, `Monetary` (Harcamalar) metriğini verir.

### 3. Sütun İsimlerini Düzenleme

```python
rfm.columns = ['Recency', 'Frequency', 'Monetary']
```

Bu satır, `agg` işlemi sonucunda oluşturulan sütun isimlerini daha anlamlı hale getirir. Böylece DataFrame’in sütunları sırasıyla 'Recency', 'Frequency', ve 'Monetary' olarak adlandırılır.

### 4. Negatif ve Sıfır Harcamaları Temizleme

```python
rfm = rfm[rfm['Monetary'] > 0]
```

- **`rfm[rfm['Monetary'] > 0]`**: `Monetary` (Harcamalar) değeri 0 veya negatif olan müşterileri filtreler ve temizler. Harcamalar 0 veya negatif olan müşteriler genellikle hatalı veriler veya potansiyel olarak problemli müşterilerdir ve bu yüzden analize dahil edilmezler.

### 5. Sonuçların Görüntülenmesi

```python
print(rfm.head())
```

- **`print(rfm.head())`**: RFM analizi sonucunda elde edilen ilk birkaç müşteri verisini görüntüler. Bu, veri setindeki müşteri segmentlerinin bir örneğini incelemek için kullanılır.

### RFM Kavramları

1. **Recency (Son Alışveriş)**: Müşterinin en son alışveriş yapma tarihinden itibaren geçen gün sayısıdır. Kısa süre önce alışveriş yapan müşteriler genellikle daha aktif ve değerli kabul edilir.

2. **Frequency (Sıklık)**: Belirli bir zaman diliminde müşteri tarafından yapılan toplam alışveriş sayısıdır. Daha sık alışveriş yapan müşteriler, marka ile daha güçlü bir ilişkiye sahip olabilirler.

3. **Monetary (Harcamalar)**: Müşterinin belirli bir zaman diliminde yaptığı toplam harcama miktarıdır. Daha yüksek harcamalar yapan müşteriler, genellikle daha değerli müşteriler olarak kabul edilir.

RFM analizi, müşteri segmentasyonu ve hedeflenmiş pazarlama stratejileri geliştirmek için oldukça yararlı bir yöntemdir.

---------------------------------

Tabii, örneklerle açıklayalım. Bu kod parçası, müşteri verilerinde "Recency" (Son Alışveriş) metriğini hesaplamak için kullanılır. Recency, müşterinin en son alışveriş yaptığı tarihten bugüne kadar geçen gün sayısını ölçer. Bu metriği hesaplamak için kullanılan kodu ve işleyişi adım adım açıklayalım:

### Örnek Senaryo

Diyelim ki veri setimizde iki müşteri var ve alışveriş tarihleri aşağıdaki gibidir:

**Müşteri 1:**
- Alışveriş Tarihleri: 2024-01-10, 2024-02-05, 2024-03-01

**Müşteri 2:**
- Alışveriş Tarihleri: 2024-01-15, 2024-02-20

Ve **snapshot_date** (kesme tarihi) olarak 2024-03-10'u kabul edelim.

### Recency Hesaplama Adımları

#### 1. En Son Alışveriş Tarihini Bulma

Öncelikle, her müşteri için en son alışveriş tarihini bulmamız gerekiyor. Bu tarihler:

- **Müşteri 1** için en son alışveriş tarihi: 2024-03-01
- **Müşteri 2** için en son alışveriş tarihi: 2024-02-20

#### 2. `snapshot_date` ile Farkı Hesaplama

Daha sonra, `snapshot_date` (2024-03-10) ile bu en son alışveriş tarihleri arasındaki farkı hesaplamalıyız. Bu fark, gün cinsinden Recency değerini verecektir:

- **Müşteri 1**:
  - Son Alışveriş Tarihi: 2024-03-01
  - Gün Farkı: `2024-03-10` - `2024-03-01` = 9 gün
  - Recency: 9 gün

- **Müşteri 2**:
  - Son Alışveriş Tarihi: 2024-02-20
  - Gün Farkı: `2024-03-10` - `2024-02-20` = 19 gün
  - Recency: 19 gün

### Kodun Açıklaması

Kod parçası şu şekilde:

```python
'InvoiceDate': lambda x: (snapshot_date - x.max()).days
```

- **`x`**: Her müşterinin alışveriş tarihlerini içeren bir Pandas Series nesnesidir.
- **`x.max()`**: Bu, her müşteri için en son alışveriş tarihini döndürür.
- **`snapshot_date - x.max()`**: Bu, en son alışveriş tarihi ile kesme tarihi arasındaki farkı hesaplar. Sonuç, bir `timedelta` (zaman farkı) nesnesidir.
- **`.days`**: `timedelta` nesnesinden günü sayısını alır. Böylece, Recency gün cinsinden hesaplanmış olur.

### Özet

Bu kod parçası, her müşterinin en son alışveriş yaptığı tarihten itibaren gün sayısını hesaplayarak, müşterinin "Recency" (Son Alışveriş) değerini belirler. Kısacası, bir müşterinin en son alışveriş yapma tarihinden bu yana geçen gün sayısını hesaplayarak, müşterinin ne kadar süredir aktif olmadığını ölçer. Bu, müşteri segmentasyonunda önemli bir metriğe dönüşür.

--------------------

*  Recency: Müşterinin en son ne zaman alışveriş yaptığı (bugüne olan uzaklığı).
* Frequency: Müşterinin kaç kez alışveriş yaptığı.
* Monetary: Müşterinin toplam harcaması.

## 5. Model Kurulumu

````python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans

# Veriyi ölçeklendirme
scaler = StandardScaler()
rfm_scaled = scaler.fit_transform(rfm)

# K-Means modeli kurma
kmeans = KMeans(n_clusters=4, random_state=42)
rfm['Cluster'] = kmeans.fit_predict(rfm_scaled)

# Küme sonuçlarına bakma
print(rfm.groupby('Cluster').mean())
````

Bu kod, **K-Means** kümeleme algoritmasını kullanarak RFM (Recency, Frequency, Monetary) metrikleri üzerinde müşteri segmentasyonu yapar. İşte kodun adım adım açıklaması:

### 1. Veri Ölçeklendirme

```python
from sklearn.preprocessing import StandardScaler

# Veriyi ölçeklendirme
scaler = StandardScaler()
rfm_scaled = scaler.fit_transform(rfm)
```

- **StandardScaler**: Verileri standardize eder. Standardizasyon, verilerin ortalama değerini 0 ve standart sapmasını 1 yaparak verileri ölçeklendirir.
- **fit_transform**: `fit` yöntemi verinin istatistiklerini öğrenir (ortalama ve standart sapma) ve `transform` yöntemi bu istatistikleri kullanarak veriyi ölçeklendirir. Sonuçta her bir özelliğin (Recency, Frequency, Monetary) ortalaması 0, standart sapması 1 olur.
  
  Bu adım, verilerin farklı ölçeklerde olmaması için önemlidir. Örneğin, Recency gün sayısı (genellikle küçük bir sayı) ve Monetary (genellikle daha büyük bir sayı) arasında fark olabilir. Ölçeklendirme, tüm özelliklerin eşit ağırlıkta değerlendirilmesini sağlar.

### 2. K-Means Modeli Kurma

```python
from sklearn.cluster import KMeans

# K-Means modeli kurma
kmeans = KMeans(n_clusters=4, random_state=42)
rfm['Cluster'] = kmeans.fit_predict(rfm_scaled)
```

- **KMeans**: K-Means algoritmasını kullanarak veriyi kümelemek için bir model oluşturur. `n_clusters` parametresi, kaç küme oluşturulacağını belirtir (bu örnekte 4).
- **fit_predict**: Modeli veriye uygular (fit) ve her verinin hangi kümeye ait olduğunu tahmin eder (predict). Sonuçta, her müşteri için bir küme etiketi atanır.
- **rfm['Cluster']**: Her müşteriye ait küme etiketi, RFM veri çerçevesinde yeni bir 'Cluster' sütununa eklenir.

### 3. Küme Sonuçlarına Bakma

```python
# Küme sonuçlarına bakma
print(rfm.groupby('Cluster').mean())
```

- **groupby('Cluster').mean()**: Her bir küme için Recency, Frequency ve Monetary metriklerinin ortalamalarını hesaplar. Bu, her kümenin genel özelliklerini anlamanıza yardımcı olur.
- **print**: Kümeleme sonuçlarını ekrana yazdırır. Bu çıktıyı inceleyerek farklı kümelerin ortalama Recency, Frequency ve Monetary değerlerini görebilir ve kümeler arasındaki farkları analiz edebilirsiniz.

### Özet

Bu kod, müşterileri RFM metriklerine göre 4 kümeye ayırır ve her kümenin ortalama özelliklerini hesaplar. Sonuç olarak:

- **Kümeleme**: Müşteriler benzer alışveriş davranışlarına göre gruplandırılır.
- **Ölçeklendirme**: Özelliklerin eşit ağırlıkta değerlendirilmesini sağlar.
- **Sonuçlar**: Her kümenin ortalama Recency, Frequency ve Monetary değerlerini gösterir, böylece kümelerin karakteristik özelliklerini anlamanızı sağlar.

Bu tür bir segmentasyon, pazarlama stratejilerini kişiselleştirmek ve müşteri ilişkilerini iyileştirmek için kullanılabilir.

Bu tür bir analiz, müşteri segmentasyonu yaparak pazarlama stratejilerini hedefleyebilir ve her segmentin ihtiyaçlarına uygun stratejiler geliştirebilir.

K-Means kümeleme sonucunda, müşterileri segmentlere ayırdıktan sonra tahmin yapmanın birkaç yolu vardır. Bu segmentler, müşteri davranışlarını anlamak ve çeşitli iş kararlarını desteklemek için kullanılabilir. İşte bu segmentlerle yapılabilecek bazı tahmin işlemleri ve uygulama örnekleri:

### 1. **Segment Bazında Satış Tahmini**

Her bir müşteri segmentinin ortalama `Recency`, `Frequency`, ve `Monetary` değerlerini kullanarak, belirli bir segmentin gelecekteki satın alma davranışlarını tahmin edebilirsiniz. Örneğin, bir segmentin ortalama harcama miktarı biliniyorsa, bu segmentin gelecekteki toplam satışlarını tahmin edebilirsiniz.

### 2. **Hedeflenmiş Pazarlama Kampanyaları**

Segmentlere göre özelleştirilmiş pazarlama stratejileri geliştirilebilir:
- **Yüksek Frekansta Alışveriş Yapan Müşteriler**: Bu segmentin müşteri memnuniyetini artırmak ve sadakatlerini ödüllendirmek için özel teklifler sunabilirsiniz.
- **Yüksek Harcama Yapan Müşteriler**: Bu müşterilere daha yüksek kaliteli ürünler veya premium teklifler sunarak daha fazla satış yapabilirsiniz.
- **Düşük Harcama Yapan Müşteriler**: Bu müşterilere daha fazla teşvik veya indirim sunarak onları daha fazla alışveriş yapmaya teşvik edebilirsiniz.

### 3. **Müşteri Yaşam Boyu Değeri (CLV) Tahmini**

Müşteri segmentasyonundan elde edilen bilgilerle, her segmentin müşteri yaşam boyu değerini tahmin edebilirsiniz. CLV tahmini, segmentlerin kârlılığını değerlendirmek ve uzun vadeli stratejiler geliştirmek için kullanılabilir.

### 4. **Segment Bazında Satın Alma Davranışları Analizi**

Her segmentin satın alma davranışlarını analiz edebilir ve gelecekteki satın alma eğilimlerini öngörebilirsiniz. Örneğin:
- **Recency**: Daha önce alışveriş yapmış müşterilerin tekrar alışveriş yapma olasılığını tahmin edebilirsiniz.
- **Frequency**: Belirli bir segmentin alışveriş sıklığına göre gelecekteki alışveriş trendlerini analiz edebilirsiniz.
- **Monetary**: Her segmentin harcama alışkanlıklarına göre, potansiyel yüksek değerli müşterileri belirleyebilirsiniz.

### 5. **Segmentlerin Performansını İzleme**

Zaman içinde her segmentin performansını izleyebilir ve değişiklikleri gözlemleyebilirsiniz. Örneğin, belirli bir segmentin harcama davranışları zamanla değişiyorsa, buna göre stratejileri güncelleyebilirsiniz.

### Örnek: Yüksek Harcama Segmentinin Tahmini

Farz edelim ki segmentlerin ortalama `Monetary` değerlerine göre yüksek harcama yapan bir segment tespit ettik. Bu segmentin müşterilerini hedef alarak:
- **Özelleştirilmiş Teklifler**: Yüksek harcama yapan müşterilere özel teklifler veya indirimler sunarak daha fazla satış yapmayı hedefleyebilirsiniz.
- **Satış Tahminleri**: Geçmiş harcamalardan yola çıkarak, bu segmentin gelecekteki toplam harcama miktarını tahmin edebilirsiniz. 

Özetle, K-Means segmentasyonu, müşterilerinizi gruplandırdıktan sonra, bu segmentler üzerinde tahminler yaparak pazarlama, satış stratejileri ve müşteri ilişkileri yönetimi gibi alanlarda etkili kararlar almanıza yardımcı olabilir.
